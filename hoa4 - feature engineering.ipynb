```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, StandardScaler, MinMaxScaler, RobustScaler, MaxAbsScaler, OneHotEncoder
from sklearn.feature_selection import mutual_info_classif
import scipy.stats as st
```

(1) Import 'employee.csv' menjadi DataFrame. Drop kolom yang dianggap tidak diperlukan dalam pembuatan model klasifikasi. Subset kolom 'Attrition' dalam variabel y dan Subset kolom lainnya dalam variabel X.

```python
employee = pd.read_csv('employee.csv')
employee
```

```python
#mendrop kolom EmployeeNumber karena kolom tersebut tidak bermakna
employee = employee.drop('EmployeeNumber', axis=1)
#mendrop kolom dengan value yang seragam
for col in employee.columns:
    if len(employee[col].unique()) == 1:
        employee.drop(col, axis=1, inplace=True)
        print(f"{col} has been dropped")
employee
```

```python
X = employee[employee.columns[employee.columns != 'Attrition']]
y = employee['Attrition']
```

```python
X
```

```python
y
```

(2) Split X dan y menjadi X_train, X_test, y_train, y_test dengan test size 0.2 dan pastikan rasio label pada y_train dan y_test seimbang.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
#memastikan rasio label pada y_train dan y_test seimbang
#rasio yes dan no pada y_train
y_train.value_counts()/len(y_train)
```

```python
#rasio yes dan no pada y_test
y_test.value_counts()/len(y_test)
```

Rasio terbukti sama

(3) Pada kolom numerik X_train lakukan scaling (scaler bebas)

```python
#memisahkan kolom objek dan numerik pada X_train
X_train_num = X_train.select_dtypes('number')
X_train_obj = X_train.select_dtypes('object')
```

```python
#mengecek apakah ada missing value pada X_train_num
all(X_train_num.isna().sum() == 0)
```

Karena tidak ditemukan missing value pada X_train_num maka data tersebut dapat langsung discaling dengan salah satu metode, di sini dipilih MinMax Scaling

```python
minmax = MinMaxScaler()
X_train_num_scaled = pd.DataFrame(minmax.fit_transform(X_train_num), columns=minmax.feature_names_in_)
X_train_num_scaled
```

(4) Pada kolom kategorikal X_train lakukan one hot encoding pada parameter handle_unknown beri argumen 'ignore' agar encoder robust terhadap adanya kategori di testing set yang tidak ada di training set.

```python
X_train_obj.isna().sum()
```

```python
#menyiapkan transformer untuk one hot encoding kolom kategorik
ohe = OneHotEncoder(handle_unknown='ignore')
X_train_obj_ohe = pd.DataFrame(ohe.fit_transform(X_train_obj).toarray(), columns=ohe.get_feature_names_out(X_train_obj.columns))
X_train_obj_ohe
```

(5) Gabungkan (menggunakan fungsi pd.concat()) hasil transformasi kolom numerik dan kategorikal X_train

```python
X_train_scaled = pd.concat([X_train_num_scaled, X_train_obj_ohe], axis=1)
X_train_scaled
```

(7) Lakukan feature selection menggunakan multicolinearity reduction (tentukan threshold sendiri) dan mutual information. Jumlah fitur hasil seleksi bebas.

```python
#menyeleksi fitur dengan multicollinearity reduction agar fitur-fitur yang akan digunakan pada model independen satu sama lain
drop = [] #list untuk menyimpan pasangan fitur yang berkorelasi tinggi
for col_1 in X_train_scaled.columns:
    for col_2 in X_train_scaled.columns:
        if col_1 != col_2:
            coef, pvalue = st.spearmanr(X_train_scaled[col_1], X_train_scaled[col_2])
            if pvalue < 0.05 and (coef < -0.8 or coef > 0.8):
                drop.append(col_1)
                drop.append(col_2)
```

```python
unique_drop = list(set(drop))
unique_drop
```

```python
#drop fitur dengan multicollinearity reduction
X_train_scaled_mulred = X_train_scaled.drop(unique_drop, axis=1)
X_train_scaled_mulred
```

```python
X_train_scaled
```

```python
mi = mutual_info_classif(X_train_scaled_mulred, y_train, random_state=42)
mi_features = pd.DataFrame({'features': X_train_scaled_mulred.columns, 'mutual_info': mi}).sort_values(by='mutual_info', ascending=False)
mi_features
```

```python
drop_mi = [] #list untuk menyimpan kolom yang akan didrop
for feature, mi in zip(mi_features['features'], mi_features['mutual_info']):
    if mi < 0.01:
        drop_mi.append(feature)
```

```python
drop_mi
```

```python
#drop fitur berdasarkan nilai mutual_info
X_train_scaled_mulred_mi = X_train_scaled_mulred.drop(drop_mi, axis=1)
X_train_scaled_mulred_mi
```
