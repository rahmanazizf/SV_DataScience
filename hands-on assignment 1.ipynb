(5) Buat 2 array x dan y, masing-masing memiliki 10 bilangan bulat yang di-generate menggunakan fungsi number generator dari library numpy. Buat scatter plot dengan menggunakan array x sebagai nilai axis x dan array y sebagai nilai axis y. Tambahkan title, label x, dan label y (bebas).

(6) Import covid.csv sebagai dataframe. Lakukan grouping dengan menggunakan kolom 'Country/Region' dan 'Date' sebagai acuan, dan gunakan .sum() sebagai metode agregasi. Lakukan reset index. Gunakan fungsi pd.to_datetime() untuk mengubah tipe data pada kolom 'Date'. 
<p>
Buat lineplot untuk data kasus aktiv (data kolom 'Active') untuk negara US, United Kingdom, Italy, India, Malaysia, Singapore, dan Indonesia. Jadikan data tanggal sebagai nilai axis x, jadikan data kasus harian sebagai nilai axis y. (1 negara diwakili 1 line) 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


covid = pd.DataFrame(pd.read_csv('covid.csv'))
covid_grouped = covid.groupby(['Country/Region', 'Date']).sum().reset_index()
covid_grouped['Date'] = pd.to_datetime(covid_grouped['Date'])

countries = covid_grouped['Country/Region'].isin(['US', 'United Kingdom', 'Italy', 'India', 'Malaysia', 'Singapore', 'Indonesia'])
covid_grouped = covid_grouped[countries]
covid_grouped

sns.lineplot(data = covid_grouped, x='Date', y='Active', hue='Country/Region')
plt.show()
```
