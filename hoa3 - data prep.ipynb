2. gabungkan employee_a.csv dan employee_b.csv menggunakan fungsi merge dengan menggunakan kolom 'EmployeeNumber' sebagai acuan dan gunakan argumen 'outer' untuk parameter how.

```python
#menyiapkan data employee yang akan diimputasi
employee_imp = employee.copy()
```

```python

for col in employee_imp.columns[employee.isna().sum() != 0]:
    ratio = employee_imp[col].isna().sum()/len(employee_imp[col])
    print(f"ratio missing value in {col}: {ratio}")
    if ratio <= 0.1 and employee_imp[col].dtype == 'O':
        simpleimp_mode = SimpleImputer(strategy='most_frequent')
        employee_imp[col] = simpleimp_mode.fit_transform(employee_imp[[col]])
        print(f"Simple imputation to {col} with most_frequent strategy done")
    elif ratio <= 0.1 and (employee_imp[col].dtype == 'float64' or employee_imp[col].dtype == 'int64'):
        simpleimp_median = SimpleImputer(strategy='median')
        employee_imp[col] = simpleimp_median.fit_transform(employee_imp[[col]])
        print(f"Simple imputation to {col} with median strategy done")
    elif ratio > 0.1 and ratio <= 0.3:
        print(f"KNNImputation is needed for handling missing value in {col}")
    else:
        employee_imp.drop(col, axis=1)
        print(f"{col} has been dropped")
    print(f"current number of missing value in {col}: {employee_imp[col].isna().sum()}")
    print("--"*50)
```

```python
#melakukan ordinal encoding untuk kolom bertipe objek
object_cols = [col for col in employee_imp.columns[employee_imp.dtypes == 'object']]
ord_encoding = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)
employee_imp_obj = employee_imp[object_cols].copy()
employee_imp_obj_encoded = ord_encoding.fit_transform(employee_imp_obj)
employee_imp_obj_encoded
```

```python
#mengganti nilai pada data bertipe objek dengan nilai hasil encoding
for col_num, col_name in enumerate(object_cols):
    employee_imp[col_name] = employee_imp_obj_encoded[:, col_num]
#menampilkan employee_imp yang sudah diencode
employee_imp
```

```python
#imputasi Kernel Nearest Neighbours (KNN) untuk data MaritalStatus dan Age
knnimputer = KNNImputer()
employee_imp2 = pd.DataFrame(knnimputer.fit_transform(employee_imp), columns=knnimputer.feature_names_in_)
employee_imp2
```

```python
employee_imp2.isna().sum()
```

```python
#mentrasformasi kembali kolom yang sebelumnnya bertipe data objek
employee_obj_inversed = ord_encoding.inverse_transform(employee_imp2[object_cols])
for col_num, col_name in enumerate(object_cols):
    employee_imp2[col_name] = employee_obj_inversed[:, col_num]
employee_imp2
```

```python
all(employee_imp2.isna().sum() == 0)
```

5. Untuk tiap kolom numerik yang dilakukan imputasi buatlah histogram dan kdeplot pre dan post imputation. Print mean, median, dan standar deviasi pre dan post imputation pada kolom terebut.

```python
num_cols = [col for col in employee_imp2.columns[~(employee_imp2.dtypes == 'object')]] #kolom numerik
for count, num_col in enumerate(num_cols):
    print(count+1, num_col)
    sns.histplot(data=employee_imp2, x=num_col, kde=True, bins=10, color='red', label='after')
    sns.histplot(data=employee, x=num_col, kde=True, bins=10, label='before')
    plt.legend()
    plt.show()
    print(f"mean {num_col} sebelum imputasi: {np.mean(employee[num_col])}")
    print(f"mean {num_col} setelah imputasi: {np.mean(employee_imp2[num_col])}")
    print(f"median {num_col} sebelum imputasi: {np.nanpercentile(employee[num_col], 50)}")
    print(f"median {num_col} setelah imputasi: {np.percentile(employee_imp2[num_col], 50)}")
    print(f"standar deviasi {num_col} sebelum imputasi: {employee[num_col].std()}")
    print(f"standar deviasi {num_col} setelah imputasi: {employee_imp2[num_col].std()}")
    print("*"*60)
```

```python
for count, obj_col in enumerate(object_cols):
    print(count+1, obj_col)
    sns.countplot(data = employee_imp2, y=obj_col, label='after', color='red')
    sns.countplot(data = employee, y=obj_col, label='before')
    plt.legend()
    plt.show()
    after_imp = employee_imp2[obj_col].value_counts()*100/len(employee_imp2)
    before_imp = employee[obj_col].value_counts()*100/len(employee)
    print(f"rasio unique value pre-imputation pada kolom {obj_col}:")
    print(before_imp)
    print(f"rasio unique value post-imputation pada kolom {obj_col}:")
    print(after_imp)
    print("="*60)
```
